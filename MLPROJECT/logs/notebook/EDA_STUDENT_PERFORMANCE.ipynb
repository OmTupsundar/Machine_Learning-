```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:/Users/omtup/Downloads/StudentsPerformance (1).csv")
df.head()
```

```python
df.describe()
df_label = df.copy()
df_label.isnull().sum()
```

```python
print(df_label.describe())
print(df_label.duplicated())
```

```python
df_label.head()
```

```python
print("Categories in 'Gender' variable: ",end=" ")
print(df_label['gender'].unique())

print("Categories in 'race/ethnicity' variable: ",end=" ")
print(df_label['race/ethnicity'].unique())

print("Categories in 'parental level of education' variable: ",end=" ")
print(df_label['parental level of education'].unique())

print("Categories in 'lunch' variable: ",end=" ")
print(df_label['lunch'].unique())

print("Categories in 'test preparation course' variable: ",end=" ")
print(df_label['test preparation course'].unique())
```

```python
# define numeric and categories column
numeric_features = [feature for feature in df_label.columns if df_label[feature].dtype !='O']  
categorical_features = [feature for feature in df_label.columns if df_label[feature].dtype =='O'] 
# diffrantiate btw the numeric and categoric columns
print('We have {} numerical features : {}' .format(len(numeric_features),numeric_features))
print('We have {} categorical features : {}' .format(len(categorical_features),categorical_features))
df_label.head(2)
```

```python
# print columns
df_label['total score'] = df_label['math score'] + df_label['reading score'] + df_label['writing score']
df_label['average'] = df_label['total score']/3
df_label.head(2)
```

```python
reading_less_20 = df_label[df_label['reading score'] <= 20]['average'].count()
writing_less_20 = df_label[df_label['writing score'] <= 20]['average'].count()
math_less_20 = df_label[df_label['math score'] <= 20]['average'].count()
print(f"Number of sudent with less than 20 marks in math :{math_less_20}")
print(f"Number of sudent with less than 20 marks in Writing :{writing_less_20}")
print(f"Number of sudent with less than 20 marks in Reading :{reading_less_20}")
```

## Histogram and KDE

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig ,axs = plt.subplots(1 ,3,figsize=(15,7))
plt.subplot(121)
sns.histplot(data = df_label , x = 'average', bins = 30 , kde = True ,color='g')
plt.subplot(122)
sns.histplot(data=df_label , x = 'average' , kde = True ,hue='gender')
plt.show()
```
